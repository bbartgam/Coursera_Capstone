```python
#scrape the web and foramt the data
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = urllib.request.urlopen(url)
soup = BeautifulSoup(page)
soup.prettify()
##maeke the df
table = soup.find_all('table')
df = pd.read_html(str(table))[0]
```

```python
!pip3 install lxml
```

```python
#get rid of the NA data in Borough
dropNAS=df[df['Borough']=='Not assigned'].index
df.drop(dropNAS, inplace=True)
#check our data. are there any NAs?
print(df.isna().sum())#check if there are any NAs in Neighborhoods
df=df.reset_index()

print(df.shape)
```

```python
df=df.drop(['index'], axis=1)
print(df.head())
```

```python
#get the csv files
df_loc=pd.read_csv('http://cocl.us/Geospatial_data')
df_loc.columns
```

```python
#join the data frames on Postal Code
df=pd.merge(df, df_loc, on='Postal Code')
df.head(15)
```

```python
#check to see if my data corralates to the exmaples
print(df.loc[df['Postal Code']=='M5G'])
print(df.dtypes)
```

```python
address = 'Toronto, ON, Canada'

geolocator = Nominatim(user_agent='Canada_Explorer')
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronoto are {}, {}.'.format(latitude, longitude))
```

```python
#making a list and checking it twice
locations = df[['Latitude', 'Longitude']]
locations_as_list = locations.values.tolist()
len(locations_as_list)
```

```python
to_map = folium.Map(location=[latitude, longitude], zoom_start=10)

for i in range(0, len(locations_as_list)):
    folium.CircleMarker(locations_as_list[i], popup=df['Postal Code'][i], radius = 5, color = 'green', fill = True, fill_color = 'green').add_to(to_map)
```

##Here I look at Downtown vs. not Downtwon Toronto

```python
toronto = df[df['Borough'] == 'Downtown Toronto']
toronto.reset_index(drop=True, inplace = True)
toronto.head()
```

```python
not_toronto = df[df['Borough'] != 'Downtown Toronto']
not_toronto.reset_index(drop=True, inplace=True)
not_toronto.shape
```

```python
to_map = folium.Map(location=[latitude, longitude], zoom_start=10)
x = toronto
y = not_toronto


for dataass, lng, borough, neighborhood in zip(x['Latitude'], x['Longitude'], x['Borough'], x['Neighborhood']):
 label = '{}, {}'.format(neighborhood, borough)
 label = folium.Popup(label, parse_html=True)
 folium.CircleMarker(
 [dataass, lng],
 radius=5,
 popup=label,
 color='green',
 fill=True,
 fill_color='green',
 parse_html=False).add_to(to_map)

for dataass, lng, borough, neighborhood in zip(y['Latitude'], y['Longitude'], y['Borough'], y['Neighborhood']):
 label = '{}, {}'.format(neighborhood, borough)
 label = folium.Popup(label, parse_html=True)
 folium.CircleMarker(
 [dataass, lng],
 radius=5,
 popup=label,
 color='red',
 fill=True,
 fill_color='red',
 parse_html=False).add_to(to_map)
```

##here I use K means to cluster the Neighborhoods

```python
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
k=5
cluster=df.drop(['Postal Code','Borough','Neighborhood'],1)
kmeans=KMeans(n_clusters=k, random_state=0).fit(cluster)
kmeans.labels_
df.insert(0,'cluster labels',kmeans.labels_)
```

```python
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]df.head()
```

```python
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
```

```python
map_cluster = folium.Map(location=[latitude, longitude], zoom_start=10)

for lat, lng, cluster, neighborhood in zip(df['Latitude'], df['Longitude'], df['cluster labels'], df['Neighborhood']):
 label = '{}, {}, {}'.format(neighborhood, borough, cluster)
 label = folium.Popup(label, parse_html=True)
 folium.CircleMarker(
 [lat, lng],
 radius=5,
 popup=label,
 color=rainbow[cluster-1],
 fill=True,
 fill_color=rainbow[cluster-1],
 parse_html=False).add_to(map_cluster)
map_cluster
```
